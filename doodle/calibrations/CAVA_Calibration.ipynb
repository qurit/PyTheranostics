```python
from doodle.calibrations.gamma_camera import GammaCamera
import matplotlib.pyplot as plt #for visualization and plots
import numpy as np

%load_ext autoreload
%autoreload 2
```

```python
isotope = 'Lu177'
```

```python
site_id = 'CAVA'
```

```python
db_dic = {
    'db_file':'/mnt/y/Sara/PR21_enrolment/PR21_DB.xlsx',
    'sheet_names':['calibration_data'],
    'header':0,
}
```

```python
# SymbiaT is the one without SerialNumber in the DICOM header
symbia_cal_file = '/mnt/y/Sara/PR21_enrolment/CAVA/Symbia/source_c/1.2.826.0.1.3680043.10.740.2702333679528003325003027076748824814.dcm'

# Intevo is the one with serial number in DICOM header
intevo_cal_file = '/mnt/y/Sara/PR21_enrolment/CAVA/Intevo/source_c/1.2.826.0.1.3680043.10.740.3006661208617321729325567106081169954.dcm'
```

## SymbiaT Sensitivity Calculation

```python
SymbiaT = GammaCamera(isotope=isotope,dicomfile=symbia_cal_file,db_dic=db_dic,cal_type='planar')
```

```python
SymbiaT.db_df['cal_data']
```

```python
SymbiaT.get_sensitivity(site_id = site_id)
```

```python
SymbiaT.calfactor_to_database(site_id = site_id)
```

## Intevo Sensitivity Calculation

```python
Intevo = GammaCamera(isotope=isotope,dicomfile=intevo_cal_file,db_dic=db_dic,cal_type='planar')
```

```python
Intevo.db_df['cal_data']
```

```python
Intevo.get_sensitivity(site_id = site_id)
```

```python
Intevo.calfactor_to_database(site_id = site_id)
```
