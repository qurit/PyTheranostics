### M2: Extrapolation of TIAC_H from TIAC_A applying a relative mass scaling factor (rMSF)

```python
all_organ = list(TIAC_A.index)
all_organ.remove('Tumor')
all_organ.remove('Tail')
```

```python
human.phantom_mass.rename(index={'Left Colon': 'Large Intestine'}, inplace=True)
human.phantom_mass.rename(index={'Bone Surfaces': 'Skeleton'}, inplace=True)
human.phantom_mass.rename(index={'Heart Contents': 'Blood'}, inplace=True)
```

```python
mouse_mass = 25
```

```python
mouse.not_inphantom_notumor.remove('Tail')
mouse.not_inphantom_notumor
```

```python
human.not_inphantom_notumor.remove('Tail')
human.not_inphantom_notumor.remove('Tumor')
```

```python
in_phantom_organs = [x for x in all_organ if x not in human.not_inphantom_notumor]
```

```python
in_phantom_organs
```

```python
human.disintegrations=pd.DataFrame(index=in_phantom_organs,columns=['h Female', 'h Male'])
for org in in_phantom_organs:
    print(org)
    if org == 'Remainder Body':
        mouse_rof = mouse.literature_mass['25g'].loc[mouse.not_inphantom_notumor].sum()
    elif org in mouse.phantom_mass.index:
        mouse_mass_org = mouse.phantom_mass.loc[org, '25g']
    elif org in mouse.literature_mass.index:
        mouse_mass_org = mouse.literature_mass.loc[org, '25g']
    else:
        print(f'Dont have a mass for {org}')
        
    if org == 'Remainder Body':
        human_rof_female = human.literature_mass['Female'].loc[human.not_inphantom_notumor].sum()
        human_rof_male = human.literature_mass['Male'].loc[human.not_inphantom_notumor].sum()
    elif org in human.phantom_mass.index:
        human_mass_org_female = human.phantom_mass['Female'].loc[org]
        human_mass_org_male = human.phantom_mass['Male'].loc[org]
        
    human.disintegrations.loc[org, 'h Female']=TIAC_A.loc[org, '%ID*h']*(mouse_mass/human.phantom_mass['Female'].loc['Body'])*(human_mass_org_female/mouse_mass_org)
    human.disintegrations.loc[org, 'h Male']=TIAC_A.loc[org, '%ID*h']*(mouse_mass/human.phantom_mass['Male'].loc['Body'])*(human_mass_org_male/mouse_mass_org)
    if org == 'Remainder Body':
        human.disintegrations.loc['Remainder Body', 'h Female'] = TIAC_A['%ID*h'].loc[human.not_inphantom_notumor].sum()*(mouse_mass/human.phantom_mass['Female'].loc['Body'])*(human_rof_female/mouse_rof)
        human.disintegrations.loc['Remainder Body', 'h Male'] = TIAC_A['%ID*h'].loc[human.not_inphantom_notumor].sum()*(mouse_mass/human.phantom_mass['Male'].loc['Body'])*(human_rof_female/mouse_rof)
```

```python

human.disintegrations.rename(index={'Skeleton':'Bone Surfaces'}, inplace=True)
human.disintegrations.rename(index={'Large Intestine':'Left Colon'}, inplace=True)
human.disintegrations.rename(index={'Blood':'Heart Contents'}, inplace=True)
#### I took Large Intestine TIAC from mouse. But now my source organ is divided in 3 pieces: left colon, right colon and rectum.

human.disintegrations.loc['Red Marrow'] = 0.34 * human.disintegrations.loc['Heart Contents']
human.disintegrations.loc['Right Colon'] = human.disintegrations.loc['Left Colon']
human.disintegrations.loc['Rectum', 'h Female'] = (70/145) * human.disintegrations.loc['Left Colon', 'h Female'] #multiply by mass of rectum divided by mass of Colon
human.disintegrations.loc['Rectum', 'h Male'] = (70/150) * human.disintegrations.loc['Left Colon', 'h Male']
human.disintegrations.sort_index(inplace=True)
human.disintegrations
```
