```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/mnt/y/Sara/Preclinical_dosimetry/prova/M1.csv', skiprows=10, nrows=26, usecols=['Target Organ', 'Total', 'ICRP-103 ED'])
df.set_index('Target Organ', inplace=True)
df['Method'] = 'M1'
df['Isotope'] = '177Lu'
df['Tracer'] = 'SB03178'
df.loc['Effective Dose', 'ICRP-103 ED'] = df['ICRP-103 ED'].sum()

m1 = df
```

```python
df = pd.read_csv('/mnt/y/Sara/Preclinical_dosimetry/prova/M2.csv', skiprows=10, nrows=26, usecols=['Target Organ', 'Total', 'ICRP-103 ED'])
df.set_index('Target Organ', inplace=True)
df['Method'] = 'M2'
df['Isotope'] = '177Lu'
df['Tracer'] = 'SB03178'
df.loc['Effective Dose', 'ICRP-103 ED'] = df['ICRP-103 ED'].sum()

m2 = df
```

```python
df = pd.read_csv('/mnt/y/Sara/Preclinical_dosimetry/prova/M3.csv', skiprows=10, nrows=26, usecols=['Target Organ', 'Total', 'ICRP-103 ED'])
df.set_index('Target Organ', inplace=True)
df['Method'] = 'M3'
df['Isotope'] = '177Lu'
df['Tracer'] = 'SB03178'
df.loc['Effective Dose', 'ICRP-103 ED'] = df['ICRP-103 ED'].sum()

m3 = df
m3
```

```python
# Merge the two DataFrames on 'Target Organ'
merged_df = pd.merge(pd.merge(m1, m2, on='Target Organ'), m3, on='Target Organ')
merged_df
merged_df = merged_df.drop('Total Body')
merged_df = merged_df.drop('Effective Dose')
```

```python
merged_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming merged_df is your DataFrame

# Plotting
width = 0.2  # Width of the bars
ind = np.arange(len(merged_df))

fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(ind - width, merged_df['Total_x'], width, label='M1')
bar2 = ax.bar(ind, merged_df['Total_y'], width, label='M2')
bar3 = ax.bar(ind + width, merged_df['Total'], width, label='M3')

# Add labels, title, and legend
ax.set_xlabel('Target Organ')
ax.set_ylabel('AD (mGy/MBq)')
ax.set_title('Comparison of Method 1, 2 and 3')
ax.set_xticks(ind)
ax.set_xticklabels(merged_df.index, rotation=45, ha='right')
ax.legend()
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

# Show the plot
plt.show()
```
